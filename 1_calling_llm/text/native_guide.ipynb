## OpenAI Responses API
OpenAI's latest response API, recommended for new builds. Supports message-style input with richer features; other providers may not accept this format yet.

- **Pros**: Cutting-edge features; structured responses.  
- **Cons**: OpenAI-only; newer API may have bugs.  
- **Use Case**: Advanced OpenAI integrations.
- **Documentation Link**: [OpenAI](https://platform.openai.com/docs)

```python
# imports
import os
from dotenv import load_dotenv
load_dotenv(override=True)

from openai import OpenAI

# Connect to OpenAI client library
openai = OpenAI()

# prepare message
message = "Hello! This is my first ever message to you!"
messages = [{"role": "user", "content": message}]

# simple test call to openai using Responses API
openai_response = openai.responses.create(model="gpt-4o", input=messages)
print("\n# openai response:")
print(openai_response.output_text)
```

## OpenAI Direct Requests API
Uses the requests library to make raw HTTP calls to OpenAI, bypassing the client SDK.

- **Pros**: Full HTTP control; no SDK dependencies.  
- **Cons**: Manual error handling; verbose code.  
- **Use Case**: Custom integrations or low-level debugging.
- **Documentation Link**: [OpenAI](https://platform.openai.com/docs)

```python
# imports
import os
from dotenv import load_dotenv
load_dotenv(override=True)
import requests

# raw HTTP requests don't read the environment, so grab the key manually
openai_api_key = os.getenv('OPENAI_API_KEY')

# prepare message
message = "Hello! This is my first ever message to you!"
messages = [{"role": "user", "content": message}]

# simple test call to openai via requests
url = "https://api.openai.com/v1/chat/completions"
headers = {
    "Authorization": f"Bearer {openai_api_key}",
    "Content-Type": "application/json"
}
data = {
    "model": "gpt-4o",
    "messages": messages
}
response = requests.post(url, headers=headers, json=data)

# print results
result = response.json()
print("\n# openai response:")
print(result['choices'][0]['message']['content'])
```

## Gemini SDK
Uses Google's official generativeai SDK for direct access to the latest Gemini models (e.g., Gemini 3.1 Flash-Lite).

- **Pros**: Direct access; multimodal support.  
- **Cons**: Google-only; requires API key.  
- **Use Case**: Gemini-specific features like vision.
- **Documentation Link**: [Gemini](https://ai.google.dev/gemini-api/docs)

```python
# imports

import os
from dotenv import load_dotenv
load_dotenv(override=True)
import google.generativeai as genai

# Configure the API key
# the genai library requires an explicit API key string; it doesn't
# automatically pull from the environment, hence the lookup here.
google_api_key = os.getenv('GOOGLE_API_KEY')
genai.configure(api_key=google_api_key)

# prepare message
message = "Hello! This is my first ever message to you!"

# simple test call to gemini
model = genai.GenerativeModel('gemini-3.1-flash-lite-preview')
response = model.generate_content(message)
print("\n# gemini response:")
print(response.text)
```

## Anthropic SDK
Uses Anthropic's official SDK for direct access to Claude models.

- **Pros**: High-quality, safety-focused responses; strong reasoning.  
- **Cons**: Anthropic-only; requires API key.  
- **Use Case**: Advanced reasoning and analysis.
- **Documentation Link**: [Anthropic](https://docs.anthropic.com/)

```python
# imports
import os
from dotenv import load_dotenv
load_dotenv(override=True)
import anthropic

# Connect to Anthropic client
client = anthropic.Anthropic()

# prepare message
message = "Hello! This is my first ever message to you!"

# simple test call to claude
response = client.messages.create(
    model="claude-3-5-sonnet-20241022",
    max_tokens=100,
    messages=[{"role": "user", "content": message}]
)
print("\n# claude response:")
print(response.content[0].text)
```

## Grok Direct Requests
Uses the requests library to make raw HTTP calls to xAI's Grok API, bypassing any client SDK.

- **Pros**: Full HTTP control; no SDK dependencies.  
- **Cons**: Manual error handling; verbose code.  
- **Use Case**: Custom integrations with Grok.
- **Documentation Link**: [xAI](https://docs.x.ai/)

```python
# imports
import os
from dotenv import load_dotenv
load_dotenv(override=True)
import requests

# raw HTTP requests don't read the environment, so grab the key manually
xai_api_key = os.getenv('XAI_API_KEY')

# prepare message
message = "Hello! This is my first ever message to you!"
messages = [{"role": "user", "content": message}]

# simple test call to grok via requests
url = "https://api.x.ai/v1/chat/completions"
headers = {
    "Authorization": f"Bearer {xai_api_key}",
    "Content-Type": "application/json"
}
data = {
    "model": "grok-4",
    "messages": messages
}
response = requests.post(url, headers=headers, json=data)

# print results
result = response.json()
print("\n# grok response:")
print(result['choices'][0]['message']['content'])
```

## Mistral SDK
Uses Mistral AI's official SDK for direct access to Mistral models.

- **Pros**: Efficient, multilingual support; strong performance.  
- **Cons**: Mistral-only; requires API key.  
- **Use Case**: Fast inference and multilingual tasks.
- **Documentation Link**: [Mistral](https://docs.mistral.ai/)

```python
# imports
import os
from dotenv import load_dotenv
load_dotenv(override=True)
from mistralai import Mistral

# Connect to Mistral client
client = Mistral(api_key=os.getenv('MISTRAL_API_KEY'))

# prepare message
message = "Hello! This is my first ever message to you!"

# simple test call to mistral
response = client.chat.complete(
    model="mistral-large-latest",
    messages=[{"role": "user", "content": message}]
)
print("\n# mistral response:")
print(response.choices[0].message.content)
```

## Cohere SDK
Uses Cohere's official SDK for direct access to Cohere models.

- **Pros**: Strong text generation and embeddings; user-friendly.  
- **Cons**: Cohere-only; requires API key.  
- **Use Case**: Text generation, summarization, and embeddings.
- **Documentation Link**: [Cohere](https://docs.cohere.com/)

```python
# imports
import os
from dotenv import load_dotenv
load_dotenv(override=True)
import cohere

# Connect to Cohere client
co = cohere.Client(api_key=os.getenv('COHERE_API_KEY'))

# prepare message
message = "Hello! This is my first ever message to you!"

# simple test call to cohere
response = co.generate(
    model="command-r7b-12-2024",
    prompt=message,
    max_tokens=100
)
print("\n# cohere response:")
print(response.generations[0].text)
```

## HuggingFace Transformers
Uses HuggingFace Transformers for direct local model loading and inference. Ideal for open-source models without API calls. Requires local hardware/GPU for large models.

- **Pros**: Free, offline, customizable.  
- **Cons**: Hardware-intensive; manual prompt handling.  
- **Use Case**: Local experimentation with open-source LLMs.
- **Documentation Link**: [HuggingFace](https://huggingface.co/docs/transformers).

```python
# imports
from transformers import pipeline

# prepare message (converted to prompt for transformers)
message = "Hello! This is my first ever message to you!"
prompt = f"User: {message}\nAssistant:"

# simple test call to huggingface model
generator = pipeline("text-generation", model="microsoft/DialoGPT-small")
response = generator(prompt, max_new_tokens=50, truncation=True, num_return_sequences=1)
print("\n# huggingface response:")
print(response[0]['generated_text'].replace(prompt, "").strip())
```
