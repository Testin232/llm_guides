# Using LiteLLM Python Library
Examples for non-streaming and streaming calls.


## Setup

```python
# imports
from dotenv import load_dotenv
load_dotenv(override=True)

import litellm

# local helper
from streaming_utils import get_chunk_content
```

```python
# model
model_name = "ollama/llama3.2:1b"

# prepare message
system_message = "You are a helpful assistant."
user_message = "Tell me a joke about programming."

# LiteLLM uses a list of dicts for messages
messages = [{"role": "system", "content": system_message}, {"role": "user", "content": user_message}]
```

## Non-Streaming

```python
def message(messages, model, **kwargs):
    used_kwargs = dict(kwargs)

    response = litellm.completion(model=model, messages=messages, **used_kwargs)
    return response.choices[0].message.content

# simple test call
print(f"Using model: {model_name}")
response = message(messages, model_name)
print(response)
```

## Streaming

```python
def stream_message(messages, model_name, **kwargs):
    used_kwargs = dict(kwargs)

    stream = litellm.completion(model=model_name, messages=messages, stream=True, **used_kwargs)
    full_response = ""
    for chunk in stream:
        content = get_chunk_content(chunk)
        print(content, end="", flush=True)
        full_response += content
    return full_response

# simple test call
print(f"Using model: {model_name}")
response = stream_message(messages, model_name)
```
