## accelerate

Accelerate enables PyTorch code to run across distributed configurations with minimal changes. It handles distributed training and inference.

**Main features:**
- Easy distributed training
- Support for DeepSpeed, FSDP
- Mixed-precision training

```python
from trl import SFTTrainer
from transformers import TrainingArguments

# Note: This is a basic setup; actual training requires data and more config
# For demo, just import and show trainer class
print("SFTTrainer imported successfully")
```

## trl

TRL (Transformer Reinforcement Learning) is a library for training transformer language models with methods like Supervised Fine-Tuning (SFT), DPO, GRPO, etc.

**Main features:**
- Trainers for online/offline RL methods
- Reward modeling
- Knowledge distillation

```python
from peft import LoraConfig, get_peft_model
from transformers import AutoModelForSequenceClassification, AutoTokenizer

# Load a base model
model_name = "distilbert-base-uncased"
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Apply PEFT
peft_config = LoraConfig(task_type="SEQ_CLS", r=8, lora_alpha=32, lora_dropout=0.1, target_modules=["q_lin", "k_lin", "v_lin", "out_lin"])
model = get_peft_model(model, peft_config)
print("PEFT model loaded")
```

## peft

🤗 PEFT (Parameter-Efficient Fine-Tuning) enables adapting large pretrained models to downstream tasks without fine-tuning all parameters, reducing costs.

**Main features:**
- Methods like LoRA for efficient fine-tuning
- Integration with Transformers and Accelerate
- Comparable performance to full fine-tuning

```python
from transformers import pipeline

# Load a sentiment analysis pipeline
classifier = pipeline("sentiment-analysis")
result = classifier("I love using Hugging Face!")
print(result)
```

## transformers

Transformers is the core library for state-of-the-art machine learning models in text, vision, audio, and multimodal tasks. It provides model definitions for inference and training.

**Main features:**
- Pipeline for easy inference
- Trainer for training
- Support for over 1M+ model checkpoints

```python
from datasets import load_dataset

# Load a sample dataset
dataset = load_dataset("imdb", split="train[:1%]")  # Small subset for demo
print(dataset[0])
```

## datasets

🤗 Datasets is a library for easily accessing and sharing AI datasets for various tasks. It supports loading datasets in a single line of code and provides powerful data processing methods.

**Main features:**
- Load datasets from Hugging Face Hub
- Process and stream large datasets efficiently
- Integration with Apache Arrow for performance

```python
from huggingface_hub import HfApi

# Initialize API
api = HfApi()

# List some public models (example)
models = api.list_models(limit=5)
for model in models:
    print(model.id)
```

## huggingface_hub

The `huggingface_hub` library allows you to interact with the Hugging Face Hub, a platform for sharing pre-trained models and datasets. It provides a simple way to download files, create repositories, and upload content.

**Main features:**
- Download and upload models/datasets
- Manage repositories
- Run inference on hosted models
