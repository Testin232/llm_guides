# Comparing vLLM and Ollama

This notebook compares vLLM with Ollama for local LLM inference, noting that they use different model formats.

## Key Differences

- **Ollama**: Easy setup, quantized models (GGUF), CPU-friendly, CLI/API.
- **vLLM**: High performance, GPU-optimized, Hugging Face models, better for scaling.

## When to Use Which

- Use Ollama for simplicity and your existing models.
- Use vLLM for speed and advanced features.

To compare performance, run similar prompts in both and time them.
