# Streaming Text Generation with vLLM

This notebook demonstrates how to stream text generation in real-time using vLLM, useful for chatbots or live responses.

```python
from vllm import LLM, SamplingParams
```

```python
llm = LLM(model="microsoft/DialoGPT-small")
```

## Streaming Generation

vLLM can stream tokens. For simple demo, we'll generate and print token by token.

```python
import time

sampling_params = SamplingParams(max_tokens=20)
outputs = llm.generate("Tell me a story:", sampling_params)

# Simulate streaming by printing word by word
text = outputs[0].outputs[0].text
words = text.split()
for word in words:
    print(word, end=' ', flush=True)
    time.sleep(0.1)  # Simulate delay
print()
```
