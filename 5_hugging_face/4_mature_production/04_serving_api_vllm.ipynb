# Serving vLLM as an API

This notebook shows how to serve a vLLM model as a local API for integration with apps.

## Running the vLLM Server

In terminal, run:
```bash
python -m vllm.entrypoints.openai.api_server --model microsoft/DialoGPT-small
```

Then, use requests to query it.

````python
import requests

response = requests.post("http://localhost:8000/v1/chat/completions", json={
    "model": "microsoft/DialoGPT-small",
    "messages": [{"role": "user", "content": "Hello!"}]
})
print(response.json())
````
