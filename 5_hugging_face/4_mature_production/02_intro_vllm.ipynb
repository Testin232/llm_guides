# Introduction to vLLM: Efficient LLM Inference

This notebook introduces vLLM, a library for fast and memory-efficient inference of large language models (LLMs). We'll cover installation, basic usage, and simple use cases, with notes on your uv environment and Ollama models.

## What is vLLM?
vLLM is an open-source library that optimizes LLM inference using techniques like continuous batching and paged attention. It's ideal for local, GPU-accelerated text generation, especially for production apps needing speed and scalability.

## When to Use vLLM
- For fast, local inference on GPUs.
- When handling multiple concurrent requests.
- Compared to Ollama: vLLM is more performant for large models but requires Hugging Face format (not Ollama's GGUF). If you have Ollama models, you can convert them or use vLLM with HF equivalents.

## 2. Import Required Libraries

````python
from vllm import LLM, SamplingParams
import time
````

## 3. Load a Large Language Model

Load a model from Hugging Face. Use your HF token for gated models.

````python
# Load a small model for demo (replace with your choice)
llm = LLM(model="microsoft/DialoGPT-small")  # No token needed for this public model
````

## 4. Generate Text with Basic Parameters

````python
prompt = "Hello, how are you?"
outputs = llm.generate(prompt)
for output in outputs:
    print(output.outputs[0].text)
````

## 5. Customize Sampling Parameters

````python
sampling_params = SamplingParams(temperature=0.7, top_p=0.9, max_tokens=50)
outputs = llm.generate(prompt, sampling_params)
print(outputs[0].outputs[0].text)
````

## 6. Handle Multiple Prompts (Batching)

````python
prompts = ["Hello!", "How's the weather?", "Tell me a joke."]
outputs = llm.generate(prompts, sampling_params)
for i, output in enumerate(outputs):
    print(f"Prompt {i+1}: {output.outputs[0].text}")
````

## 7. Measure Inference Speed

````python
start = time.time()
outputs = llm.generate(prompt, sampling_params)
end = time.time()
print(f"Time taken: {end - start:.2f} seconds")
````

## 8. Compare with Standard Hugging Face Transformers

Install transformers if needed: `uv add transformers torch`

Then compare speed.

````python
# Comparison code (run separately)
from transformers import AutoModelForCausalLM, AutoTokenizer
model = AutoModelForCausalLM.from_pretrained("microsoft/DialoGPT-small")
tokenizer = AutoTokenizer.from_pretrained("microsoft/DialoGPT-small")
inputs = tokenizer(prompt, return_tensors="pt")
start = time.time()
outputs = model.generate(**inputs, max_new_tokens=50)
end = time.time()
print(f"Transformers time: {end - start:.2f} seconds")
# Compare with vLLM's time above
````
